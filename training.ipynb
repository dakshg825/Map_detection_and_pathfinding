```python
#datasets for taining the keras model to identify the map
train_ds= tf.keras.utils.image_dataset_from_directory(
    directory='/kaggle/input/the-blind-fight/train',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
validation_ds= tf.keras.utils.image_dataset_from_directory(
    directory='/kaggle/input/the-blind-fight/validation',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
#normalizing the image colours
def process(image, label):
  image= tf.cast(image/255, tf.float32)
  return image, label

train_ds= train_ds.map(process)
validation_ds= validation_ds.map(process)
```

```python
#The keras model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),padding='valid',strides=2))

model.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),padding='valid',strides=2))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(3, activation="softmax"))
```

```python
model.summary()
```

```python
#compiling the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#fitting the model
history=model.fit(train_ds, epochs=10, validation_data=validation_ds)
```

```python
#save the model
model.save("/kaggle/working/map_recognition.keras")
```
